```python
name = pd.DataFrame(df,columns=['product_name'])

name.product_name = name.product_name.fillna (0)
name.product_name = name.product_name.astype ('string')
name.dtypes
```

**TRANSFORMANDO TIPO DE DATOS**

```python
Dpeso = pd.DataFrame(df,columns=['product_information'])
Dpeso['product_information'] = Dpeso.product_information.str.slice(18,50)
Dpeso = Dpeso[:20].copy()
Dpeso

df['product_information'] = Dpeso
```

```python
Fabricante = pd.DataFrame(df,columns=['manufacturer'])
Fabricante.manufacturer = Fabricante.manufacturer.astype ('string')
```

```python
Categoria = pd.DataFrame(df,columns=['amazon_category_and_sub_category'])
Categoria['amazon_category_and_sub_category'] = Categoria.amazon_category_and_sub_category.str.slice(10,38)
Categoria = Categoria[:20].copy()
Categoria.amazon_category_and_sub_category = Categoria.amazon_category_and_sub_category.astype ('string')

df['amazon_category_and_sub_category'] = Categoria
```

```python
Dprice = pd.DataFrame(df,columns=['price'])
Dprice['price'] = Dprice.price.str.replace(',','.')
Dprice['price'] = Dprice.price.str.replace('-','')
Dprice['price'] = Dprice.price.str.slice(1,6)
Dprice.loc[Dprice['price'].isnull(), 'price'] = 0

Dprice = Dprice[:10].copy()
Dprice

Dprice.price = Dprice.price.fillna (0)
Dprice.price = Dprice.price.astype (float)

df['price'] = Dprice
```

```python
Vnumber = pd.DataFrame(df,columns=['number_of_reviews'])
Vnumber['number_of_reviews'] = Vnumber.number_of_reviews.str.replace(',','.')

Vnumber
Vnumber.number_of_reviews = Vnumber.number_of_reviews.fillna (0)
Vnumber.number_of_reviews = Vnumber.number_of_reviews.astype (float)
df['number_of_reviews'] = Vnumber
```

```python
Rating = pd.DataFrame(df,columns=['average_review_rating'])
Rating['average_review_rating'] = Rating.average_review_rating.str.slice(0,3)

Rating.average_review_rating = Rating.average_review_rating.fillna (0)
Rating.average_review_rating = Rating.average_review_rating.astype (float)

Rating = Rating[:50].copy()
Rating

df['average_review_rating'] = Rating
```

```python
stock  = pd.DataFrame(df,columns=['number_available_in_stock'])
stock['number_available_in_stock'] = stock.number_available_in_stock.str.slice(0,1)
stock.loc[stock['number_available_in_stock'].isnull(), 'number_available_in_stock'] = 0
stock = stock[:10].copy()
stock

stock.number_available_in_stock = stock.number_available_in_stock.fillna (0)
stock.number_available_in_stock = stock.number_available_in_stock.astype (float)

df['number_available_in_stock'] = stock
```

**GRAFICA DEL MAYOR FABRICANTE** 
Tops 50 fabricantes 

```python
#EL MAYOR FABRICANTE 

dfManufacturer = pd.crosstab(df['manufacturer'], columns = 'count').sort_values(by ="count", ascending=False)
graf = dfManufacturer[:50].copy()
graf

graf.plot(kind = 'bar', 
             stacked = 'True',         
             alpha = 0.5,              
             width = 0.9,              
             figsize=(18,4));            
```
